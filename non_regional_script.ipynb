```python
def dataframe_value_validation(df1, col1, df2, col2):
    def find_df_name(df):
        name = [name for name, obj in globals().items() if id(obj) == id(df)]
        return name[0] if name else None
    
    try:
        df1_name = find_df_name(df1)
        df2_name = find_df_name(df2)
        value_1 = round(df1[col1].sum(),2)
        value_2 = round(df2[col2].sum(),2)
        variance = abs(value_1)-abs(value_2)
        
        if value_1 == value_2:
            return(f"The specified columns within the following dataframes:"+ " "+str(df1_name).upper()+" & " + str(df2_name).upper()+ ", " +"are equal with a value of: "+ str(value_1))
        else:
            return("The Datasets are not equal with a variance of "+ str(variance)+"... "+ str(df1_name).upper()+ f" = ${value_1} and"+" "+str(df2_name).upper()+ f" = ${value_2}")
            
    except KeyError:
        return("One or both of the specified columns do not exist in the dataframes.")
```

```python
def coep_KPI_validation_check(mergeddf, primary_column_in_merged_name ,amount_merged_col_name, settlement_df_name, settlement_df_amount_name):
    
    def find_df_name(df):
        name = [name for name, obj in globals().items() if id(obj) == id(df)]
        return name[0] if name else None
     
    df1_name = find_df_name(mergeddf)
    df2_name = find_df_name(settlement_df_name)
    
    ##DF1 is merged
    ##DF2 is 
    settlement_row_count = len(settlement_df_name[settlement_df_amount_name])
    empty_row_count_after_merge = len(mergeddf[primary_column_in_merged_name].isnull())
    percentage_rows_empty = (empty_row_count_after_merge / settlement_row_count) * 100
    settlement_dollar_amount = settlement_df_name[settlement_df_amount_name].sum()
    empty_dollar_amount_after_merge = mergeddf.loc[mergeddf[primary_column_in_merged_name].isnull(), amount_merged_col_name].sum()
    percentage_amount_unaccounted_for = ((1 - (empty_dollar_amount_after_merge / settlement_dollar_amount)) * 100)
    
    
    return f"""Comparison of: {df1_name} against the originally filtered Settlement DataFrame: {df2_name}, we have the following results:
    The count of settlements in the {df2_name} dataframe is {settlement_row_count} and our merged dataframe has {empty_row_count_after_merge} empty rows, resulting in a {percentage_rows_empty}% unmatched settlements.
    The Total dollar amount of the {df2_name} is ${settlement_dollar_amount} dollars. There is ${empty_dollar_amount_after_merge} dollars that are unaccountd for 
    for in the merged dataframe {df1_name}, resultin in {percentage_amount_unaccounted_for}% of total dollars unaccounted for"""
    
    


```
